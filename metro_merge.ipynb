```python
import re
def replace_name(name_station):
    name_station = name_station.lower().replace("ё","е")
    delete_sub_strings = {" МЦК", " Г/З", " М/З", " КС", " БКЛ", " Р", " К", " (КАЛУЖСКО-РИЖСКАЯ)", 
                          " (СОКОЛЬНИЧЕСКАЯ)", " АП", " Арбатско-Покровский", " Калининско-Солнцевский",
                          " (арбатско-покровская линия)", " мкц", "вход-выход 1 в южный вестибюл",
                          "вход-выход 2 в северный вестибюль", " мр", " (калининская)", "(арбатско покровская линия)",
                          " (калужско рижская)", " (сокольническая)", "вход выход 1 в южный вестибюль", " (калининская)",
                          "вход выход в вестибюль (нкр)", "вход-выход в вестибюль (нкр)", "вход-выход 1 в южный вестибюль"
                         }
    for sub_str in delete_sub_strings:
        sub_str = sub_str.replace("(", "\(").replace(")", "\)")
        
#         name_station = re.sub(r"" + sub_str.lower() + r"$" ,"", name_station)
        name_station = re.sub(r"{}$".format(sub_str.lower()) ,"", name_station)
        
    name_station = name_station.replace(" пл.", " площадь")
    name_station = name_station.replace("ул.", "улица")
    name_station = name_station.replace("б-р", "бульвар")
    name_station = name_station.replace("ввц", "вднх")
    name_station = name_station.replace("-", " ")
    name_station = name_station.replace("ак.", "академика")
    name_station = name_station.replace("улица сергея эйзенштейна", "улица эйзенштейна")
    name_station = name_station.replace("выставочный центр", "вднх")
    name_station = name_station.replace("варшавская", "каширская")
    name_station = name_station.replace("тверская", "пушкинская")
    name_station = name_station.strip()
    
    
    
    
    
    
    
    
    return name_station
```

```python

join_metro = pd.merge(df_metro_data_mos_without_duplicate[["StationName", "AdmArea", "District", "Longitude_WGS84", "Latitude_WGS84", "VestibuleType", "NameOfStation"]], df_pass_metro_06_09_22[["StationName" , "PERIOD" , "Total"]], on="StationName", how="left")
# train_df = pd.merge(train_source_df, train_result_df[["Параметр" ,"mark"]], on="Параметр", how="left").fillna(outliers_value).rename(columns={\"Параметр\":\"Time\"})\n",
#     "train_df[\"mark\"] = train_df[\"mark\"].astype(int)\n",


# join_metro.loc[join_metro[join_metro["PERIOD"]=="п1 (4-10)"].index,"Period_before_10"] = join_metro[join_metro["PERIOD"]=="п1 (4-10)"]["Total"]
# join_metro.loc[join_metro[join_metro["PERIOD"]=="п2 (10-16)"].index,"Period_10_16"] = join_metro[join_metro["PERIOD"]=="п2 (10-16)"]["Total"]
# join_metro.loc[join_metro[join_metro["PERIOD"]=="п3 (16-04)"].index,"Period_after_16"] = join_metro[join_metro["PERIOD"]=="п3 (16-04)"]["Total"]


join_metro = join_metro.pivot_table(index=["StationName", "AdmArea", "District", "Longitude_WGS84", "Latitude_WGS84", "VestibuleType", "NameOfStation"], columns="PERIOD",
                     values="Total", aggfunc='first').reset_index()
join_metro = join_metro.rename(columns={"п1 (4-10)": "Pass_before_10", "п2 (10-16)": "Pass_10_16", "п3 (16-04)":"Pass_after_16"})
join_metro.index.name = "ID"
join_metro
```

```python
# Объединение данных по станциям
union_stations ={
    "арбатская": ["арбатская", "боровицкая", "александровский сад"],
    "красногвардейская": ["красногвардейская", "зябликово"],
    "новокузнецкая": ["новокузнецкая","третьяковская"],
    "охотный ряд": ["охотный ряд","площадь революции"],
    "чеховская": ["чеховская","пушкинская"],
    "динамо": ["динамо","петровский парк"],
    "каховская": ["каховская","севастопольская"],
    "добрынинская": ["добрынинская","серпуховская"],
    "цветной бульвар": ["цветной бульвар","трубная"],
    "менделеевская": ["менделеевская","новослободская"],
    "савеловская": ["савеловская",""],
    "новоясеневская": ["новоясеневская","битцевский парк"],
    "воронцовская": ["воронцовская","калужская"],
    "площадь гагарина": ["площадь гагарина","ленинский проспект"],
    "тургеневская": ["тургеневская","сретенский бульвар","чистые пруды"],
    "лубянка": ["лубянка","кузнецкий мост"],
    "черкизовская": ["черкизовская","локомотив"],
    "выставочная": ["выставочная","деловой центр", "международная"],
    "чкаловская": ["чкаловская","курская"],
    "хорошевская": ["хорошевская","полежаевская"],
    "краснопресненская": ["краснопресненская","баррикадная"],
    "таганская": ["таганская","марксистская"],
    "крестьянская застава": ["крестьянская застава","пролетарская"],
    "косино": ["косино","лермонтовский проспект"],
    "римская": ["римская","площадь ильича"],
}
for key in union_stations:
    join_metro.loc[join_metro[join_metro["StationName"] == key].index,["Pass_before_10"]] = join_metro[join_metro["StationName"].isin(union_stations[key])]["Pass_before_10"].sum()
    join_metro.loc[join_metro[join_metro["StationName"] == key].index,["Pass_10_16"]] = join_metro[join_metro["StationName"].isin(union_stations[key])]["Pass_10_16"].sum()
    join_metro.loc[join_metro[join_metro["StationName"] == key].index,["Pass_after_16"]] = join_metro[join_metro["StationName"].isin(union_stations[key])]["Pass_after_16"].sum()

    join_metro = join_metro.drop(join_metro[join_metro["StationName"].isin(set(union_stations[key]) - {key})].index)
join_metro.shape
```

```python
join_metro["PassTotal"] = join_metro["Pass_before_10"] + join_metro["Pass_10_16"] + join_metro["Pass_after_16"]
```
